## NumPy Solutions

### Task 1: Create a Linear Dataset

**Solution:**

```python
# Task 1 Solution
feature = np.arange(6, 21)  # Create array from 6 to 20 (inclusive)
label = 3 * feature + 4      # Calculate: 3 * feature + 4

print("Feature:", feature)
print("Label:", label)
```

**Explanation:**
- `np.arange(6, 21)` creates integers from 6 to 20 (21 is exclusive)
- Broadcasting applies `3 * feature + 4` to each element

### Task 2: Add Noise to the Dataset

**Solution:**

```python
# Task 2 Solution
noise = np.random.random(len(label)) * 4 - 2  # Random values in [-2, +2]
noisy_label = label + noise                    # Add noise to label

print("Original label:", label)
print("Noise:", noise)
print("Noisy label:", noisy_label)
```

**Explanation:**
- `np.random.random()` generates values in [0, 1)
- Multiply by 4 to get [0, 4)
- Subtract 2 to shift to [-2, +2)
- Alternative: `np.random.uniform(-2, 2, len(label))`

### Task 3: Array Statistics

**Solution:**

```python
# Task 3 Solution
mean_value = np.mean(noisy_label)  # Calculate mean
max_value = np.max(noisy_label)    # Calculate maximum
min_value = np.min(noisy_label)    # Calculate minimum
std_value = np.std(noisy_label)    # Calculate standard deviation

print(f"Mean: {mean_value}")
print(f"Max: {max_value}")
print(f"Min: {min_value}")
print(f"Std: {std_value}")
```

**Explanation:**
- NumPy provides many statistical functions
- Alternative methods: `noisy_label.mean()`, `noisy_label.max()`, etc.

### Task 4: Create a 2D Array

**Solution:**

```python
# Task 4 Solution
zeros_matrix = np.zeros((4, 5))                    # 4x5 matrix of zeros
ones_matrix = np.ones((4, 5))                      # 4x5 matrix of ones
random_matrix = np.random.randint(10, 51, (4, 5))  # 4x5 matrix of random integers [10-50]

print("Zeros shape:", zeros_matrix.shape)
print("Ones shape:", ones_matrix.shape)
print("Random matrix:\n", random_matrix)
```

**Explanation:**
- Shape is specified as a tuple: `(rows, columns)`
- `np.random.randint(10, 51, (4, 5))` generates integers in [10, 50]

### Task 5: Array Reshaping

**Solution:**

```python
# Task 5 Solution
sequence = np.arange(1, 13)     # Create array [1, 2, 3, ..., 12]
reshaped = sequence.reshape(3, 4)  # Reshape to 3x4 matrix

print("Original:", sequence)
print("\nReshaped (3x4):\n", reshaped)
```

**Explanation:**
- `np.arange(1, 13)` creates [1, 2, ..., 12]
- `.reshape(3, 4)` transforms to 3 rows × 4 columns
- Total elements must match: 12 = 3 × 4

## Pandas Solutions

First, load the Titanic dataset:

```python
# Load the Titanic dataset
titanic = pd.read_csv('../datasets/titanic.csv')
print(f"Loaded {len(titanic)} rows")
titanic.head()
```

### Task 6: Filter by Gender

**Solution:**

```python
# Task 6 Solution
female_passengers = titanic[titanic['Sex'] == 'female']  # Filter for Sex == 'female'
male_passengers = titanic[titanic['Sex'] == 'male']      # Filter for Sex == 'male'

print(f"Female passengers: {len(female_passengers)}")
print(f"Male passengers: {len(male_passengers)}")
```

**Explanation:**
- `titanic['Sex'] == 'female'` creates a boolean mask
- Using the mask as index filters rows where condition is True

### Task 7: Calculate Survival Rate by Gender

**Solution:**

```python
# Task 7 Solution
female_survival_rate = (female_passengers['Survived'].sum() / len(female_passengers)) * 100
male_survival_rate = (male_passengers['Survived'].sum() / len(male_passengers)) * 100

print(f"Female survival rate: {female_survival_rate:.1f}%")
print(f"Male survival rate: {male_survival_rate:.1f}%")
```

**Explanation:**
- `.sum()` on 'Survived' column counts 1s (survivors)
- Divide by total passengers and multiply by 100 for percentage
- Alternative: `female_passengers['Survived'].mean() * 100`

### Task 8: Create a New Column

**Solution:**

```python
# Task 8 Solution
titanic['Family_Size'] = titanic['Siblings/Spouses Aboard'] + titanic['Parents/Children Aboard'] + 1

# Display result
titanic[['Name', 'Family_Size', 'Survived']].head(10)
```

**Explanation:**
- Add columns together element-wise
- `+ 1` accounts for the passenger themselves
- New column is created automatically

### Task 9: Find the Most Expensive Ticket

**Solution:**

```python
# Task 9 Solution
max_fare = titanic['Fare'].max()                # Find maximum fare
most_expensive = titanic[titanic['Fare'] == max_fare]  # Filter for rows with max fare

print(f"Maximum fare: ${max_fare:.2f}")
print("\nPassenger(s) with most expensive ticket:")
most_expensive[['Name', 'Pclass', 'Fare']]
```

**Explanation:**
- `.max()` finds the maximum value
- Boolean indexing filters for rows matching this value
- Multiple passengers may share the same maximum fare

### Task 10: Group by Class and Calculate Average Age

**Solution:**

```python
# Task 10 Solution
average_age_by_class = titanic.groupby('Pclass')['Age'].mean()

print("Average age by passenger class:")
print(average_age_by_class)
```

**Explanation:**
- `.groupby('Pclass')` groups rows by passenger class
- `['Age']` selects the Age column
- `.mean()` calculates average for each group
- Result shows 1st class passengers were older on average

### Task 11: Multiple Conditions Filter

**Solution:**

```python
# Task 11 Solution
filtered_passengers = titanic[
    (titanic['Sex'] == 'female') & 
    (titanic['Pclass'] == 1) & 
    (titanic['Age'] < 30)
]

print(f"Number of young first-class female passengers: {len(filtered_passengers)}")
print("\nFirst few:")
filtered_passengers[['Name', 'Age', 'Pclass', 'Sex']].head()
```

**Explanation:**
- Use `&` (and) to combine conditions
- Each condition must be in parentheses
- Use `|` for OR conditions, `~` for NOT
- All conditions must be True for a row to be included

### Task 12: Missing Data Analysis

**Solution:**

```python
# Task 12 Solution
missing_counts = titanic.isnull().sum()  # Count missing values per column
age_missing_percent = (titanic['Age'].isnull().sum() / len(titanic)) * 100  # Calculate percentage

print("Missing values per column:")
print(missing_counts)
print(f"\nPercentage of missing Age values: {age_missing_percent:.1f}%")
```

**Explanation:**
- `.isnull()` creates boolean mask where True = missing
- `.sum()` counts True values (missing data points)
- Age has missing values that need handling in real analysis
- Alternative: `titanic.isna().sum()`

## Bonus: Data Insights

Let's combine what we learned to gain insights:

```python
# Survival rate by class and gender
print("Survival Rate by Class and Gender:\n")
survival_by_class_gender = titanic.groupby(['Pclass', 'Sex'])['Survived'].mean() * 100
print(survival_by_class_gender)
print("\nKey finding: First-class females had the highest survival rate!")
```

```python
# Family size impact on survival
print("Survival Rate by Family Size:\n")
survival_by_family = titanic.groupby('Family_Size')['Survived'].agg(['mean', 'count'])
survival_by_family['survival_rate'] = survival_by_family['mean'] * 100
print(survival_by_family[['count', 'survival_rate']].sort_values('survival_rate', ascending=False))
print("\nKey finding: Small families (2-4 people) had better survival rates than solo travelers or large families!")
```
