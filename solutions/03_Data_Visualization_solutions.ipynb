```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set_theme()
titanic = pd.read_csv('../datasets/titanic.csv')
titanic.head()
```

### Task 1: Histogram of ages

```
sns.histplot(data=titanic, x="Age");
```

### Task 2: Did age matter for survival?

```
sns.histplot(data=titanic, x="Age", hue="Survived", multiple="stack");
```

### Task 3: Age vs. ticket price

```
sns.scatterplot(data=titanic, x="Age", y="Fare", hue="Pclass");
```

### Task 4: Pair plot

```
subset = titanic[["Age", "Fare", "Survived"]].dropna()
sns.pairplot(subset, hue="Survived");
```

### Task 5: Read the plots 🔎

*Example observations (yours may differ):*

1. Younger passengers (children) survived at a higher rate than adults — the survivor bars dominate at low ages in Task 2.
2. Higher fares / first class (Task 3) are associated with more survivors, suggesting class affected survival chances.
