```
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    data.data, data.target, test_size=0.3, random_state=42, stratify=data.target)
sc = StandardScaler()
X_train_c = sc.fit_transform(X_train_c)
X_test_c = sc.transform(X_test_c)
print('Breast cancer data ready.')
```

### Exercise: SVM Kernel Comparison

```
kernels = ['linear', 'rbf', 'poly']
for kernel in kernels:
    clf = SVC(kernel=kernel)
    clf.fit(X_train_c, y_train_c)
    acc = accuracy_score(y_test_c, clf.predict(X_test_c))
    print(f'{kernel:>7} kernel accuracy: {acc:.3f}')
print('\nOn this scaled dataset the linear and RBF kernels usually perform best (~97%).')
```
