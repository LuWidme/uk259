## NumPy Solutions

### Task 1: Create a Linear Dataset

### Solution

```python
feature = np.arange(6, 21)   # 6 to 20 inclusive (21 is exclusive)
label = 3 * feature + 4      # Apply linear formula to each element

print("Feature:", feature)
print("Label:", label)
```

**Explanation:**
- `np.arange(6, 21)` creates integers from 6 to 20 — the stop value 21 is exclusive
- NumPy **broadcasting** applies `3 * feature + 4` element-wise to the entire array in one operation

### Task 2: Add Noise to the Dataset

### Solution

```python
noise = np.random.random(len(label)) * 4 - 2  # Random values in [-2, +2]
noisy_label = label + noise

print("Original label:", label)
print("Noise:", noise)
print("Noisy label:", noisy_label)
```

**Explanation:**
- `np.random.random()` generates values in [0, 1)
- Multiply by 4 to get range [0, 4), then subtract 2 to shift to [-2, +2)
- General formula: `random * (max - min) + min`
- Alternative: `np.random.uniform(-2, 2, len(label))`

### Task 3: Array Statistics

### Solution

```python
mean_value = np.mean(noisy_label)
max_value = np.max(noisy_label)
min_value = np.min(noisy_label)
std_value = np.std(noisy_label)

print(f"Mean: {mean_value}")
print(f"Max: {max_value}")
print(f"Min: {min_value}")
print(f"Std: {std_value}")
```

**Explanation:**
- NumPy provides these as both functions (`np.mean(arr)`) and methods (`arr.mean()`)
- Standard deviation measures how spread out the values are from the mean
- These statistics are fundamental in ML for understanding your data

### Task 4: Create a 2D Array

### Solution

```python
zeros_matrix = np.zeros((4, 5))                     # 4x5 matrix of zeros
ones_matrix = np.ones((4, 5))                       # 4x5 matrix of ones
random_matrix = np.random.randint(10, 51, (4, 5))   # 4x5 matrix of random ints [10, 50]

print("Zeros shape:", zeros_matrix.shape)
print("Ones shape:", ones_matrix.shape)
print("Random matrix:\n", random_matrix)
```

**Explanation:**
- Shape is specified as a tuple: `(rows, columns)` — here `(4, 5)` means 4 rows, 5 columns
- `np.random.randint(10, 51, (4, 5))` uses 51 because the upper bound is exclusive — so we get integers from 10 to 50

### Task 5: Array Reshaping

### Solution

```python
sequence = np.arange(1, 13)        # [1, 2, 3, ..., 12]
reshaped = sequence.reshape(3, 4)  # Reshape to 3 rows x 4 columns

print("Original:", sequence)
print("\nReshaped (3x4):\n", reshaped)
```

**Explanation:**
- `np.arange(1, 13)` creates 12 elements
- `.reshape(3, 4)` rearranges into 3 rows and 4 columns — total elements must match (12 = 3 x 4)
- Elements fill row by row (row-major order)

## Pandas Solutions

First, load the Titanic dataset:

```python
titanic = pd.read_csv('../datasets/titanic.csv')
print(f"Loaded {len(titanic)} rows")
titanic.head()
```

### Task 6: Filter by Gender

### Solution

```python
female_passengers = titanic[titanic['Sex'] == 'female']
male_passengers = titanic[titanic['Sex'] == 'male']

print(f"Female passengers: {len(female_passengers)}")
print(f"Male passengers: {len(male_passengers)}")
```

**Explanation:**
- `titanic['Sex'] == 'female'` creates a boolean Series (True/False for each row)
- Using this as an index keeps only the rows where the condition is True
- This is called **boolean indexing** and is one of the most important Pandas operations

### Task 7: Calculate Survival Rate by Gender

### Solution

```python
female_survival_rate = (female_passengers['Survived'].sum() / len(female_passengers)) * 100
male_survival_rate = (male_passengers['Survived'].sum() / len(male_passengers)) * 100

print(f"Female survival rate: {female_survival_rate:.1f}%")
print(f"Male survival rate: {male_survival_rate:.1f}%")
```

**Explanation:**
- The 'Survived' column contains 1 (survived) and 0 (died), so `.sum()` counts survivors
- Dividing by total passengers gives the proportion, multiplied by 100 for a percentage
- Shorter alternative: `female_passengers['Survived'].mean() * 100` — `.mean()` on 0/1 values gives the proportion directly

### Task 8: Create a New Column

### Solution

```python
titanic['Family_Size'] = titanic['Siblings/Spouses Aboard'] + titanic['Parents/Children Aboard'] + 1

titanic[['Name', 'Family_Size', 'Survived']].head(10)
```

**Explanation:**
- Assigning to `titanic['Family_Size']` creates a new column
- Pandas adds the two columns element-wise, then adds 1 for the passenger themselves
- The new column is immediately available for further analysis

### Task 9: Find the Most Expensive Ticket

### Solution

```python
max_fare = titanic['Fare'].max()
most_expensive = titanic[titanic['Fare'] == max_fare]

print(f"Maximum fare: ${max_fare:.2f}")
print("\nPassenger(s) with most expensive ticket:")
most_expensive[['Name', 'Pclass', 'Fare']]
```

**Explanation:**
- `.max()` returns the single largest value in the column
- We then filter for all rows matching this value — there could be multiple passengers with the same fare
- Alternative: `titanic.nlargest(1, 'Fare')` returns the row(s) with the highest fare directly

### Task 10: Group by Class and Calculate Average Age

### Solution

```python
average_age_by_class = titanic.groupby('Pclass')['Age'].mean()

print("Average age by passenger class:")
print(average_age_by_class)
```

**Explanation:**
- `.groupby('Pclass')` splits the data into groups (one per class)
- `['Age']` selects the Age column within each group
- `.mean()` calculates the average for each group
- Notice that 1st class passengers tend to be older — wealthier passengers could afford better tickets

### Task 11: Multiple Conditions Filter

### Solution

```python
filtered_passengers = titanic[
    (titanic['Sex'] == 'female') &
    (titanic['Pclass'] == 1) &
    (titanic['Age'] < 30)
]

print(f"Number of young first-class female passengers: {len(filtered_passengers)}")
print("\nFirst few:")
filtered_passengers[['Name', 'Age', 'Pclass', 'Sex']].head()
```

**Explanation:**
- Combine multiple conditions with `&` (AND) — each condition must be in parentheses
- Other operators: `|` for OR, `~` for NOT
- A row is included only when **all** conditions are True

### Task 12: Missing Data Analysis

### Solution

```python
missing_counts = titanic.isnull().sum()
age_missing_percent = (titanic['Age'].isnull().sum() / len(titanic)) * 100

print("Missing values per column:")
print(missing_counts)
print(f"\nPercentage of missing Age values: {age_missing_percent:.1f}%")
```

**Explanation:**
- `.isnull()` creates a boolean DataFrame where True = missing value
- `.sum()` counts True values per column
- Missing data is very common in real datasets — handling it properly is a key part of data preprocessing (covered in later notebooks)
