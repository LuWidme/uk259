### 1. Titanic - *who survived?* (*) - Classification

The passengers of the Titanic, with age, sex, ticket class, fare, and whether they survived. A classic first project.

- **Task:** Classification - predict the `Survived` column (0 = no, 1 = yes)
- **Watch out for:** missing `Age` values; `Sex` and `Embarked` need encoding
- **Bonus question:** which feature mattered most for survival?

```python
# Load Titanic
titanic = pd.read_csv("datasets/titanic.csv")
print(titanic.shape)
titanic.head()
```

### 2. Palmer Penguins - *which species?* (*) - Classification or Clustering

Body measurements (bill length, flipper length, body mass) for three penguin species. Small, friendly, and great for plots.

- **Task (classification):** predict `species` from the measurements
- **Task (clustering):** ignore `species`, use K-Means to find groups, then check if they match the real species
- **Watch out for:** a few missing rows - drop or fill them

```python
# Load Penguins (built into seaborn)
penguins = sns.load_dataset("penguins")
print(penguins.shape)
penguins.head()
```

### 3. Breast Cancer Wisconsin - *benign or malignant?* (**) - Classification

Real diagnostic measurements from cell-nucleus images, labelled benign or malignant. A serious, real-world medical classification task - and a good reminder that ML decisions can carry weight.

- **Task:** Classification - predict the diagnosis (0 = malignant, 1 = benign)
- **Watch out for:** 30 numeric features - scaling matters (Notebook 04); think about which errors are worse (a missed cancer vs. a false alarm - recall vs. precision, Notebook 06)

```python
# Load Breast Cancer Wisconsin (built into scikit-learn)
from sklearn.datasets import load_breast_cancer
bc = load_breast_cancer(as_frame=True)
cancer = bc.frame          # features + a 'target' column (0 = malignant, 1 = benign)
print(cancer.shape)
cancer.head()
```

### 4. Diamonds - *what's it worth?* (**) - Regression

~54,000 diamonds with carat, cut, colour, clarity, and price. Predict the price from the features - the same idea as predicting a song's popularity or a movie's rating from its properties.

- **Task:** Regression - predict `price` (a number, in dollars)
- **Watch out for:** it's big (sample it if things get slow); `cut`, `color`, `clarity` are categories that need encoding
- **Tip:** start with just `carat` as a single predictor, then add more (Notebook 07's multiple regression)

```python
# Load Diamonds (built into seaborn)
diamonds = sns.load_dataset("diamonds")
print(diamonds.shape)
diamonds.head()
```

### 5. California Housing - *predict the price* (***) - Regression

Median house values across California districts, with income, house age, rooms, and location. The most feature-rich option - pick this if you want a real challenge with numeric regression.

- **Task:** Regression - predict `MedHouseVal` (median house value)
- **Watch out for:** 8 numeric features, all on different scales - scaling and feature selection really matter here

```python
# Load California Housing (built into scikit-learn)
from sklearn.datasets import fetch_california_housing
ch = fetch_california_housing(as_frame=True)
housing = ch.frame         # features + 'MedHouseVal' target
print(housing.shape)
housing.head()
```
