1. **Histogram** — how is a single column distributed?
2. **Scatter plot** — how do two columns relate to each other?
3. **Pair plot** — all scatter plots and histograms at once

## Meet the penguins 🐧

For the demos we use the **Palmer Penguins** dataset: body measurements of 344 penguins from three species (Adelie, Chinstrap, Gentoo) recorded at a research station in Antarctica. Seaborn can download it for us:

```python
penguins = sns.load_dataset("penguins")
print(f"{penguins.shape[0]} penguins, {penguins.shape[1]} columns")
penguins.head()
```

## 1. Histograms — one column at a time

A **histogram** splits the values of one column into bins and counts how many rows fall into each bin. It answers: *what values are typical? Are there groups?*

```python
sns.histplot(data=penguins, x="flipper_length_mm");
```

Interesting — there seem to be **two bumps**. Why? Let's color the bars by species using the `hue` parameter:

```python
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack");
```

Now it's obvious: Gentoo penguins simply have longer flippers than the other two species. One extra parameter (`hue`) turned a vague pattern into an explanation. Remember `hue` — it works in almost every Seaborn plot.

You can control the granularity with `bins`: try `bins=10` (coarse) vs `bins=50` (fine).

## 2. Scatter plots — two columns together

A **scatter plot** draws one dot per row: one column on the x-axis, another on the y-axis. It answers: *do these two values have something to do with each other?*

```python
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species");
```

Each species forms its own cloud. A machine learning model could use exactly these two measurements to tell the species apart — this is the core idea behind **classification**, which we'll cover on Day 3. When the dots of a group cluster together, a model can learn to recognize that group.

## 3. Pair plots — the big overview

With more than two numeric columns, checking every pair by hand gets tedious. `sns.pairplot` draws **all combinations at once**: scatter plots for every pair, histograms on the diagonal.

```python
sns.pairplot(penguins, hue="species", height=2.0);
```

One command, complete overview: you immediately see which measurement pairs separate the species well (bill length vs. bill depth) and which don't (bill depth vs. body mass, lots of overlap).

**That's all the theory you need.** Histogram, scatter plot, pair plot — with these three you can explore almost any dataset. Time to practice.
